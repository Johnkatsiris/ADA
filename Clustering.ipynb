```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans, DBSCAN
from sklearn.metrics import silhouette_score, davies_bouldin_score, calinski_harabasz_score
from sklearn.cluster import AgglomerativeClustering
import scipy.cluster.hierarchy as sch

# Step 1: Load the data
file_path = 'data.csv'  # Replace with your file path
data = pd.read_csv(file_path, delimiter=';')

# Step 2: Convert columns to numeric types
data['Sales'] = pd.to_numeric(data['Sales'].str.replace('.', '', regex=False).str.replace(',', '.', regex=False), errors='coerce')
data['Cost'] = pd.to_numeric(data['Cost'], errors='coerce')
data['Price'] = pd.to_numeric(data['Price'], errors='coerce')
data['Profit'] = pd.to_numeric(data['Profit'], errors='coerce')
data['Quantity'] = pd.to_numeric(data['Quantity'], errors='coerce')

# Step 3: Check for non-finite values and remove problematic rows
data = data.dropna(subset=['Quantity', 'Sales', 'Profit', 'Cost', 'Price'])

# Step 4: Summarize data for each customer
customer_summary = data.groupby('Customer ID').agg({
    'Quantity': 'sum',
    'Sales': 'sum',
    'Profit': 'sum',
    'Cost': 'sum',
    'Price': 'mean'
}).reset_index()

# Step 5: Feature scaling
scaler = StandardScaler()
scaled_features = scaler.fit_transform(customer_summary[['Quantity', 'Sales', 'Profit', 'Cost', 'Price']])

# Step 6: Dimensionality reduction using PCA
pca = PCA(n_components=2)
pca_features = pca.fit_transform(scaled_features)

# Step 7: Determine optimal number of clusters using the Elbow Method
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(pca_features)
    wcss.append(kmeans.inertia_)

plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), wcss, marker='o')
plt.title('Elbow Method for Optimal Number of Clusters')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.xticks(range(1, 11))
plt.grid(True)
plt.show()

# Step 8: Apply K-means clustering with optimal clusters (assume 4 based on Elbow method)
kmeans = KMeans(n_clusters=4, random_state=42)
customer_summary['KMeans_Cluster'] = kmeans.fit_predict(pca_features)

# Step 9: Apply Hierarchical clustering
hc = AgglomerativeClustering(n_clusters=4, affinity='euclidean', linkage='ward')
customer_summary['Hierarchical_Cluster'] = hc.fit_predict(pca_features)

# Step 10: Evaluate clusters
silhouette_avg = silhouette_score(pca_features, customer_summary['KMeans_Cluster'])
db_index = davies_bouldin_score(pca_features, customer_summary['KMeans_Cluster'])
ch_score = calinski_harabasz_score(pca_features, customer_summary['KMeans_Cluster'])

print(f'Silhouette Score: {silhouette_avg}')
print(f'Davies-Bouldin Index: {db_index}')
print(f'Calinski-Harabasz Index: {ch_score}')

# Step 11: Visualize clusters using pair plots
sns.pairplot(customer_summary, hue='KMeans_Cluster', vars=['Quantity', 'Sales', 'Profit', 'Cost', 'Price'])
plt.suptitle('Pair Plot for K-means Clusters', y=1.02)
plt.show()

sns.pairplot(customer_summary, hue='Hierarchical_Cluster', vars=['Quantity', 'Sales', 'Profit', 'Cost', 'Price'])
plt.suptitle('Pair Plot for Hierarchical Clusters', y=1.02)
plt.show()

# Step 12: Apply DBSCAN clustering
dbscan = DBSCAN(eps=0.5, min_samples=5)
customer_summary['DBSCAN_Cluster'] = dbscan.fit_predict(pca_features)

# Step 13: Visualize DBSCAN clusters
sns.scatterplot(x=pca_features[:, 0], y=pca_features[:, 1], hue=customer_summary['DBSCAN_Cluster'], palette='viridis')
plt.title('DBSCAN Clusters')
plt.xlabel('PCA Feature 1')
plt.ylabel('PCA Feature 2')
plt.show()

# Step 14: Summarize clusters (optional)
kmeans_summary = customer_summary.groupby('KMeans_Cluster').agg({
    'Quantity': ['mean', 'sum'],
    'Sales': ['mean', 'sum'],
    'Profit': ['mean', 'sum'],
    'Cost': ['mean', 'sum'],
    'Price': ['mean', 'sum']
}).reset_index()

hierarchical_summary = customer_summary.groupby('Hierarchical_Cluster').agg({
    'Quantity': ['mean', 'sum'],
    'Sales': ['mean', 'sum'],
    'Profit': ['mean', 'sum'],
    'Cost': ['mean', 'sum'],
    'Price': ['mean', 'sum']
}).reset_index()

dbscan_summary = customer_summary.groupby('DBSCAN_Cluster').agg({
    'Quantity': ['mean', 'sum'],
    'Sales': ['mean', 'sum'],
    'Profit': ['mean', 'sum'],
    'Cost': ['mean', 'sum'],
    'Price': ['mean', 'sum']
}).reset_index()

print("K-means Clustering Summary:")
print(kmeans_summary)

print("Hierarchical Clustering Summary:")
print(hierarchical_summary)

print("DBSCAN Clustering Summary:")
print(dbscan_summary)
```
